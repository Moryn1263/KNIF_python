# Czym jest biblioteka Pandas?

```python
import pandas as pd
import numpy as np 
```

**Pandas** (skrt od "Python data analysis"), jest jednym z najbardziej rozbudowanych pakietw, do analizy danych, w Python. Moemy za jego pomoc, wczytywa dane, czyci, modyfikowa, a nawet analizowa. Wszystko to co umoliwia nam SQL, Excel i duo wicej.

Pandas jest zbudowany na bazie dwch innych biobliotek  numpy i matplotlib.

**NumPy** to biblioteka dodajca obsug duych, wielowymiarowych tabel i macierzy.

**Matplotlib** to z kolei biblioteka do tworzenia wykresw i wizualizacji danych.

Oprcz funkcjonalnoci, Pandas jest bibliotek intuicyjn i przejrzyst.

## Pandas Series

Pierwszy typ danych to "Series". Dla analogii, moemy porwna j do kolumny z Excela. Dziaa ona podobnie do listy w Python, jednak daje nam wiksze moliwoci.

Dla przykadu stwrzmy "kolumn" liczbow:

```python
kol = pd.Series([-4,0,2,4,6,23]) 
kol
```

### Dziaania przerne

```python
# mnoenie
kol*10
```

```python
# modu
kol.abs()
```

```python
# troch statystyki
kol.describe()
```

### Manipulacja indeksami

```python
kol.index
```

```python
kol.index += 200
kol
```

```python
kol.index = ["a", "b", "c", "d", "e", "f"]
kol
```

Pandas Series moe zawiera rne typy danych, nie tylko liczbowe

Stwrzmy list, ktra zawiera wicej ni jeden typ danych:

```python
tr = ["Wyczkowski", 390, False, "Wyczkowski", None, "Baczyski", "Orzeszkowa", 21.37, np.NaN]
tr
```

```python
# Tworzymy pandasow seri danych (1-wymiarow)
IC = pd.Series(tr)
IC
```

```python
# Wywietlanie pierwszych wyrazw (domylnie 5)
IC.head(3)
```

```python
# Wywietlanie ostatnich wyrazw (domylnie 5)
IC.tail(2)
```
